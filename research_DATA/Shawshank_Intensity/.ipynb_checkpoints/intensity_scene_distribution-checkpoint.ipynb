```python
path = '/Users/sohee/Documents/research_DATA/Shawshank_Intensity/'

data = pd.read_csv('shawshank_scene_arousal.csv', encoding='utf-8')
print(data.head(10))
```

```python
andy_arousal = []

for arousal in data['andy_arousal']:
    arousal = arousal.strip('[]')
#     print(arousal)
    andy_arousal.append(arousal)

print(len(andy_arousal))
```

```python
andy_frame = []

for frame in data['Andy_frame']:
#     print(type(frame))
    frame = frame[21:26]
    andy_frame.append(frame)

print(len(andy_frame))
```

```python
from matplotlib.ticker import MaxNLocator
from collections import namedtuple
```

```python
fig,ax = plt.subplots()
ax.set_xlabel('Frame Number')
ax.set_ylabel('arousal level')
# ax.yaxis.set_major_formatter(FormatStrFormatter('%100d'))

plt.plot(andy_frame, andy_arousal)
fig_size = plt.rcParams["figure.figsize"]
print(fig_size)
fig_size[0] = 50
fig_size[1] = 5
plt.rcParams["figure.figsize"] = fig_size
plt.ylim(0,400)

# y_tick = [0,50,100,150,200,250,300,350,400]
# plt.yticks(np.arange(0, 400, step=50))
plt.title("Andy arousal level", fontsize=20)

# plt.savefig('Andy_scene_size.png')
plt.show()
```
