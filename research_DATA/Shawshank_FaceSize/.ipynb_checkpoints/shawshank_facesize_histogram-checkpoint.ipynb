```python
from matplotlib import pyplot as plt
import numpy as np
import csv
```

```python
face_size_DA = []
face_size_Fat = []
face_size_Andy = []
face_size_Brooks = []
face_size_Tommy = []
face_size_Bogs = []
face_size_Norton = []
face_size_Red = []
face_size_Heywood = []
face_size_Hadley = []
face_size_Elmo = []
face_size_Skeet = []
face_size_Floyd = []

with open('shawshank_face_size.csv') as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        face_size_DA.append(row[0])
        face_size_Fat.append(row[1])
        face_size_Andy.append(row[2])
        face_size_Brooks.append(row[3])
        face_size_Tommy.append(row[4])
        face_size_Bogs.append(row[5])
        face_size_Norton.append(row[6])
        face_size_Red.append(row[7])
        face_size_Heywood.append(row[8])
        face_size_Hadley.append(row[9])
        face_size_Elmo.append(row[10])
        face_size_Skeet.append(row[11])
        face_size_Floyd.append(row[12])

#     sales_times_raw = sales_times_raw[1:]

print(face_size_Fat)
```

```python
face_size_DA = list(filter(None, face_size_DA))
face_size_Fat = list(filter(None, face_size_Fat))
face_size_Andy = list(filter(None, face_size_Andy))
face_size_Brooks = list(filter(None, face_size_Brooks))
face_size_Tommy = list(filter(None, face_size_Tommy))
face_size_Bogs = list(filter(None, face_size_Bogs))
face_size_Norton = list(filter(None, face_size_Norton))
face_size_Red = list(filter(None, face_size_Red))
face_size_Heywood = list(filter(None, face_size_Heywood))
face_size_Hadley = list(filter(None, face_size_Hadley))
face_size_Elmo = list(filter(None, face_size_Elmo))
face_size_Skeet = list(filter(None, face_size_Skeet))
face_size_Floyd = list(filter(None, face_size_Floyd))
# fastest
```
