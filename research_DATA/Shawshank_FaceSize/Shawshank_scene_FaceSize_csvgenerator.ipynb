```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import csv
from collections import Counter
```

```python
scene_info = pd.read_csv('scene_profiles.csv')
scene_start = scene_info['scene_start'].tolist()
scene_number = scene_info['scene_number'].tolist()
print(type(scene_start))
print(scene_start)
```

```python
path = '/Users/sohee/Documents/research_DATA/Shawshank_FaceSize/'
lines = []
andy_sizes = [[0.0, 0.0] for i in range(len(scene_start))]

with open(path + 'shawshank_frames_facesize.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    for line in reader:
#         print(line)
        lines.append(line)

for row_line in lines[1:]:
#     print(row_line[2])
    try:
        frame = int(row_line[2][10:15])
        for i in range(len(scene_start)-1):
            if frame >= scene_start[i] and frame < scene_start[i+1]:
                #andy_sizes[i].append(int(row_line[3]))
                andy_sizes[i][0] += int(row_line[3])
                andy_sizes[i][1] += 1
#                 print(andy_size)
    except ValueError:
        pass
```

```python
print(andy_sizes)
```

```python
andy_sizes_mean = []
for i in range(len(andy_sizes)):
	if andy_sizes[i][1] == 0:
		val = 0
	else:
		val = andy_sizes[i][0] / andy_sizes[i][1]
	#val = np.mean(andy_sizes[i])
	andy_sizes_mean.append(val)

for number, mean in zip(scene_number, andy_sizes_mean):
	print('scene number: {}, andy mean face size for scene: {}'.format(number, mean))
for mean in andy_sizes_mean:
	print('{}'.format(mean))
```
