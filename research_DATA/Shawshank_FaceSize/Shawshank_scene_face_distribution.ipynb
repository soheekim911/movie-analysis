```python
from matplotlib import pyplot as plt
import numpy as np
import csv
```

```python
scene_number = []
scene_size_Andy = []
scene_size_Tommy = []
scene_size_Red = []
scene_size_Norton = []
scene_size_Bogs = []
scene_size_Skeet = []

with open('shawshank_scene_face.csv') as csvDataFile:
    reader = csv.DictReader(csvDataFile)
    for row in reader:
        scene_number.append(int(row['scene_number']))
        scene_size_Andy.append(float(row['Andy']))
        scene_size_Tommy.append(float(row['Tommy']))
        scene_size_Red.append(float(row['Red']))
        scene_size_Norton.append(float(row['Norton']))
        scene_size_Bogs.append(float(row['Bogs']))
        scene_size_Skeet.append(float(row['Skeet']))
```

```python
print(scene_number)
print(scene_size_Andy)
# print(scene_size_Tommy)
# print(scene_size_Red)
# print(len(scene_size_Red))
# print(scene_size_Norton)
# print(len(scene_size_Norton))
# print(scene_size_Bogs)
```

```python
from matplotlib.ticker import MaxNLocator
from collections import namedtuple
```

```python
fig,ax = plt.subplots()
ax.set_xlabel('Scene Number')
ax.set_ylabel('Average Face Size')
# ax.yaxis.set_major_formatter(FormatStrFormatter('%100d'))

plt.bar(range(len(scene_number)), (scene_size_Andy))
fig_size = plt.rcParams["figure.figsize"]
print(fig_size)
fig_size[0] = 30
fig_size[1] = 5
plt.rcParams["figure.figsize"] = fig_size
plt.ylim(0,400)

# y_tick = [0,50,100,150,200,250,300,350,400]
# plt.yticks(np.arange(0, 400, step=50))
plt.title("Andy average face size per scene", fontsize=20)

txt = 0
for size in scene_size_Andy:
    if size != 0:
        txt += 1
print(txt)
plt.text(150, 380, 'frequency = '+ str(txt)+'\n overall average = 265.04',
        horizontalalignment='center',
        verticalalignment='top',
        fontsize = 20)

plt.savefig('Andy_scene_size.png')
plt.show()
```
