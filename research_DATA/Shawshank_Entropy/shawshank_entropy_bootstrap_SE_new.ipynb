```python
def sh_entropy(p):
    if p != 0:
        return -p * log(p)
    else:
        pass
```

```python
# n : number of resampling, theta b : entropy of each resample, theta mean : entropy with original sample

def errsqr(theta, mean): 
    if theta != None:
        a = (theta-mean)**2
        return a
    else:
        return 0
```
