## Calculate entropy for each emotion. Ignored scene-dynamics and aggregated. 

```python
def sh_entropy(p):
    if p != 0:
        return -p * log(p)
    else:
        pass
        
    
```

## Calculated entropy for 'each' scene.

```python
def sh_entropy(p):
    if p != 0:
        return -p * log(p)
    else:
        return 0
```

```python
path= '/Users/sohee/Documents/research_DATA/Shawshank_FER/'
emo= pd.read_csv(path+'scene_emotion_andy.csv')
scene_number = emo['scene_number']

neutral_p = []
happy_p = []
sad_p = []
surprise_p = []
fear_p = []
disgust_p = []
anger_p = []
contempt_p = []

for i in range(len(emo)):
    total = emo['emotion_sum'].iloc[i]
    
    neutral = emo['neutral'].iloc[i]
    happy = emo['happy'].iloc[i]
    sad = emo['sad'].iloc[i]
    surprise = emo['surprise'].iloc[i]
    fear = emo['fear'].iloc[i]
    disgust = emo['disgust'].iloc[i]
    anger = emo['anger'].iloc[i]
    contempt = emo['contempt'].iloc[i]

    neutral_p.append(neutral/total)
    happy_p.append(happy/total)
    sad_p.append(sad/total)
    surprise_p.append(surprise/total)
    fear_p.append(fear/total)
    disgust_p.append(disgust/total)
    anger_p.append(anger/total)
    contempt_p.append(contempt/total)

# print(happy_p)
# print(anger_p)

f = open('scene_emotion_prob_andy.csv','w', encoding='utf-8')
writer = csv.writer(f)
writer.writerow(['scene_number','neutral','happy','sad','surprise','fear','disgust','anger','contempt'])
for x,a,b,c,d,e,f,g,h in zip(scene_number, neutral_p,happy_p, sad_p, surprise_p, fear_p, disgust_p, anger_p, contempt_p):
    print([x, a,b,c,d,e,f,g,h])
    writer.writerow([x, a,b,c,d,e,f,g,h])
    
    
    
    
    
```
