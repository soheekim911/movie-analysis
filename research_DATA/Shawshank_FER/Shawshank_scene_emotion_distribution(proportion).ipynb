```python
path = '/Volumes/Seagate Backup Plus Drive/research_DATA/Shawshank_FER/AffectNet_whole_results/'

scene_number = []
scene_len = []

andy_happy = []
bogs_happy = []
norton_happy = []
red_happy = []
brooks_happy = []
with open(path + 'happy_per_scene.csv') as data:
    reader = csv.DictReader(data)
    for row in reader:
        scene_number.append(int(row['scene_number']))
        scene_len.append(row['scene_len'])
        andy_happy.append(float(int(row['andy'])/int(row['scene_len'])))
        bogs_happy.append(float(row['bogs'])/float(row['scene_len']))
        norton_happy.append(float(row['norton'])/float(row['scene_len']))
        red_happy.append(float(row['red'])/float(row['scene_len']))
        brooks_happy.append(float(row['brooks'])/float(row['scene_len']))
```

```python
neutral_scene = pd.read_csv('neutral_per_scene.csv')
scene_profiles = pd.read_csv('scene_profiles.csv')

charac_list = ['andy','bogs','brooks','norton','red','tommy','skeet']
dct = {}
for man in charac_list:
    dct['%s_neutral'%(man)] = neutral_scene[man].tolist()
# print(dct)
scene_number = neutral_scene['scene_number'].tolist()
scene_len = scene_profiles['scene_len'].tolist()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
plt.title("Proportion of neutral ", fontsize=50)
plt.rcParams["figure.figsize"] = (50,10)

scene_start = range(0,160,1)
plt.xticks(np.arange(min(scene_start), max(scene_start)+1, 10))

ax1.set_xlabel('Scene Number', fontsize=30)
ax1.set_ylabel('Emotion Proportion', fontsize=30)
ax1.plot(scene_number, np.array(dct['andy_neutral'])/np.array(scene_len), c='red', marker="s", label='andy')
ax1.plot(scene_number, np.array(dct['bogs_neutral'])/np.array(scene_len), c='orange', marker="o", label='bogs')
ax1.plot(scene_number, np.array(dct['norton_neutral'])/np.array(scene_len), c='purple', marker="o", label='norton')
ax1.plot(scene_number, np.array(dct['red_neutral'])/np.array(scene_len), c='green', marker="o", label='red')
ax1.plot(scene_number, np.array(dct['brooks_neutral'])/np.array(scene_len), c='blue', marker="o", label='brooks')
ax1.plot(scene_number, np.array(dct['tommy_neutral'])/np.array(scene_len), c='olive', marker="o", label='tommy')
ax1.plot(scene_number, np.array(dct['skeet_neutral'])/np.array(scene_len), c='pink', marker="o", label='skeet')
plt.legend(loc='upper right',prop={'size': 25})

plt.savefig('neutral_scene_proportion.png')
plt.show()
```
