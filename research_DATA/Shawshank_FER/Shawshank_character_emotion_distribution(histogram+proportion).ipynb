```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import csv
import pdb
import os 
from __future__ import division
```

```python
andy_emo = pd.read_csv('scene_emotion_andy.csv')
scene_number = andy_emo['scene_number'].tolist()
# print(scene_number)

charac_list = ['andy','bogs','brooks','norton','red','tommy','skeet']
new_emo = andy_emo[['neutral','happy','sad','surprise','fear','disgust','anger','contempt']]
# print(new_emo)
    
fig = plt.figure()
ax1 = fig.add_subplot(111)
plt.title("Emotion of Andy ", fontsize=50)
plt.rcParams["figure.figsize"] = (50,10)

scene_start = range(0,160,1)
plt.xticks(np.arange(min(scene_start), max(scene_start)+1, 10))

ax1.set_xlabel('Scene Number', fontsize=30)
ax1.set_ylabel('Emotion Frequency', fontsize=30)
ax1.plot(scene_number, new_emo['neutral'], c='grey', marker="o", label='neutral')
ax1.plot(scene_number, new_emo['happy'], c='blue', marker="o", label='happy')
ax1.plot(scene_number, new_emo['sad'], c='green', marker="o", label='sad')
ax1.plot(scene_number, new_emo['surprise'], c='orange', marker="o", label='surprise')
ax1.plot(scene_number, new_emo['fear'], c='purple', marker="o", label='fear')
ax1.plot(scene_number, new_emo['disgust'], c='olive', marker="o", label='disgust')
ax1.plot(scene_number, new_emo['anger'], c='red', marker="o", label='anger')
ax1.plot(scene_number, new_emo['contempt'], c='pink', marker="o", label='contempt')
plt.legend(loc='upper right',prop={'size': 25})

plt.savefig('emotion_scene_andy.png')
plt.show()

# for emo in range(len(andy_emo)):
#     emo = andy_emo[emo]
#     ax1.plot(scene_number, emo, c = 'green')
#     print(emo)
```
