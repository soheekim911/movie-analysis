```python
path = '/Volumes/Seagate Backup Plus Drive/research_DATA/Shawshank_FER/'
f = open(path + 'shawshank_scene_fer.csv', 'r')
reader = csv.reader(f)

lines = []
for line in reader:
    lines.append(line)
print(lines)
scene_start = [int(line[1]) for line in lines[1:160]] 

print(scene_start)
print(len(scene_start))
```

```python
andy_fer = [[] for i in range(len(scene_start))]
print(andy_fer)
for row_line in lines[1:]: # header 제외한 rows에서 row 한줄씩 뽑기
    try:
        frame = int(row_line[2][21:26]) #andy의 해당 frame
#         print(frame)
        for i in range(len(scene_start)-1):
            if frame >= scene_start[i] and frame < scene_start[i+1]:
                andy_fer[i].append(row_line[4])    

    except ValueError:
        pass


for i in range(2):
    andy_fer[158].append('neutral')
for i in range(4):
    andy_fer[158].append('happy')

print(andy_fer)
print(len(andy_fer))
```

```python
scene_num = 0
scene_emo = []
for line in andy_fer:
    scene_num += 1
#     print(scene_num,Counter(line))
    scene_emo.append(Counter(line))

scene_num = 0
scene_row = []
for emo in scene_emo:
    scene_num += 1
    scene = [scene_num, emo['neutral'],emo['happy'],emo['sad'],emo['surprise'],
            emo['fear'],emo['disgust'],emo['anger'],emo['contempt']]
    scene_row.append(scene)


print('scene_row_len:'+ str(len(scene_row)))
print('scene_start_len:'+ str(len(scene_start)))

for i in range(len(scene_row)):
#     print(i)
    scene_row[i].insert(1,scene_start[i])

print('------------scene_row + scene_start----------')
# print(scene_row)

# print(len(scene_row))
f = open('scene_emotion_andy.csv','w', encoding='utf-8')
writer = csv.writer(f)
writer.writerow(['scene_number','scene_start','neutral','happy','sad','surprise','fear','disgust','anger','contempt'])
for scene in scene_row:
    print(scene)
    writer.writerow(scene)
```
